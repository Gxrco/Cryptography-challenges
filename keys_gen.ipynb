# Generación de llaves 

### Llaves Dinámicas (ASCII)

```python
ASCII_TABLE = (
    "\x00\x01\x02\x03\x04\x05\x06\x07\x08\x09\x0a\x0b\x0c\x0d\x0e\x0f"
    "\x10\x11\x12\x13\x14\x15\x16\x17\x18\x19\x1a\x1b\x1c\x1d\x1e\x1f"
    " !\"#$%&'()*+,-./0123456789:;<=>?@"
    "ABCDEFGHIJKLMNOPQRSTUVWXYZ[\\]^_`"
    "abcdefghijklmnopqrstuvwxyz{|}~"
    "\x7f"
)

def gen_dynamic_key(length: int) -> str:
    import random
    key = ''.join(random.choice(ASCII_TABLE) for _ in range(length))
    return key
```

```python
print(gen_dynamic_key(16))
```

### Nuevo Cypher (ASCII) Llave K tamaño fijo

```python
def gen_new_cypher(key: str) -> str:
    cypher = ''
    for i in range(len(ASCII_TABLE)):
        cypher += ASCII_TABLE[(i + ASCII_TABLE.index(key[i % len(key)])) % len(ASCII_TABLE)]
    return cypher
```

```python
print(gen_new_cypher("MYSECRETKEY"))
```

### Nuevo Cypher (ASCII) Llave K tamaño dinámico

```python
def gen_new_cypher_dynamic(length: int) -> str:
    key = gen_dynamic_key(length)
    cypher = ''
    for i in range(len(ASCII_TABLE)):
        cypher += ASCII_TABLE[(i + ASCII_TABLE.index(key[i % len(key)])) % len(ASCII_TABLE)]
    return cypher, key
```

```python
print(gen_new_cypher_dynamic(16))
```
