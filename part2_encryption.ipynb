### Número a Binario

```python
def number_to_binary(n):
    if n == 0:
        bits = "0"
    else:
        bits = ""
        while n > 0:
            bits = str(n % 2) + bits
            n //= 2

    while len(bits) < 8:
        bits = "0" + bits

    return bits
```

```python
number_to_binary(65)
```
