### Número a Binario

```python
def number_to_binary(n):
    if n == 0:
        bits = "0"
    else:
        bits = ""
        while n > 0:
            bits = str(n % 2) + bits
            n //= 2

    while len(bits) < 8:
        bits = "0" + bits

    return bits
```

```python
number_to_binary(65)
```

### ASCII a Número

```python
"""
Prompt:
Genera un string que represente los caracteres de ASCII de 0-127 en un string.
"""
ASCII_TABLE = (
    "\x00\x01\x02\x03\x04\x05\x06\x07\x08\x09\x0a\x0b\x0c\x0d\x0e\x0f"
    "\x10\x11\x12\x13\x14\x15\x16\x17\x18\x19\x1a\x1b\x1c\x1d\x1e\x1f"
    " !\"#$%&'()*+,-./0123456789:;<=>?@"
    "ABCDEFGHIJKLMNOPQRSTUVWXYZ[\\]^_`"
    "abcdefghijklmnopqrstuvwxyz{|}~"
    "\x7f"
)

def ascii_to_number(ch):
    """
    Parámetros:
    ch -> Caracter

    Recurso:
    ASCII_TABLE -> Tabla representativa de los valores a ASCII

    Método para devovler el número de la posición correspondiente para un caracter en la tabla ASCII.
    """
    if ch == "":
        return -1

    c = ch[0]

    i = 0
    while i < 128:
        if ASCII_TABLE[i] == c:
            return i
        i += 1

    return -1
```

```python
ascii_to_number("a")
```

### Binario a Base64


```python
def binary_to_base64(bin_string):
    BASE64_TABLE = (
        "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
        "abcdefghijklmnopqrstuvwxyz"
        "0123456789+/"
    )

    while len(bin_string) % 6 != 0:
        bin_string += "0"

    base64_string = ""
    for i in range(0, len(bin_string), 6):
        six_bits = bin_string[i:i+6]
        decimal_value = int(six_bits, 2)
        base64_char = BASE64_TABLE[decimal_value]
        base64_string += base64_char

    return base64_string
```

```python
print(binary_to_base64('01000001'))
```

### Binario a ASCII

```python
def binary_to_ascii(bin_string):
    while len(bin_string) % 8 != 0:
        bin_string += "0"

    ascii_string = ""
    for i in range(0, len(bin_string), 8):
        byte = bin_string[i:i+8]
        val = int(byte, 2)
        if 0 <= val < len(ASCII_TABLE):
            ascii_string += ASCII_TABLE[val]
        else:
            ascii_string += "?"
    return ascii_string
```

```python
print(binary_to_ascii('01000001'))
```

### Bas64 a ASCII

```python
def base64_to_ascii(base64_string):
    BASE64_TABLE = (
        "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
        "abcdefghijklmnopqrstuvwxyz"
        "0123456789+/"
    )

    bin_string = ""
    for ch in base64_string:
        index = BASE64_TABLE.find(ch)
        if index == -1:
            continue
        six_bits = number_to_binary(index)[-6:]
        bin_string += six_bits

    ascii_string = ""
    for i in range(0, len(bin_string), 8):
        byte = bin_string[i:i+8]
        if len(byte) < 8:
            continue
        val = int(byte, 2)
        if 0 <= val < len(ASCII_TABLE):
            ascii_string += ASCII_TABLE[val]
        else:
            ascii_string += "?"
    return ascii_string
```

```python
print(base64_to_ascii('QQ'))
```

### XOR a Binario

```python
def xor_to_binary(xor_string, key):
    bin_string = ""
    key_length = len(key)
    for i in range(len(xor_string)):
        xor_char = xor_string[i]
        key_char = key[i % key_length]
        xor_val = ascii_to_number(xor_char)
        key_val = ascii_to_number(key_char)
        bin_val = xor_val ^ key_val
        bin_string += number_to_binary(bin_val)
    return bin_string
```

```python
print(xor_to_binary("HELLO", "KEY"))
```
